```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('reviews_train.csv')

data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.pipeline import Pipeline

from sklearn.svm import SVC

text_clf = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', SVC(kernel='linear', C=0.025)),
])

#all the data is for training

X_train = data['review']
y_train = data['label']

text_clf.fit(X_train, y_train)

data_test = pd.read_csv('reviews_test.csv')

X_testv = data_test['review']

predicted = text_clf.predict(X_testv)


print('done')

print(predicted)

#calculate the score

from sklearn import metrics

print(metrics.accuracy_score(data_test['label'], predicted))

```
